## Esfuerzos Alrededor de Excavaciones Subterráneas

<p style='text-align: justify;'>La caracterización del estado de esfuerzos y desplazamientos alrededor de cavidades es una de las cuestiones más relevantes en geomecánica. Si bien en la actualidad se dispone de métodos computacionales para realizar cálculos en excavaciones de sección irregular, las soluciones analíticas para formas simplificadas son muy útiles para dilucidar aspectos generales. Estas tienen la ventaja de que permiten visualizar el comportamiento de los resultados de acuerdo a la influencia de los parámetros involucrados.
Por otra parte, la mayoría de soluciones analíticas requieren asumir un comportamiento elástico lineal del medio, mientras que para casos no lineales se precisa de expresiones más complejas (Jaeger et al., 2011).
Esta sección solo pretende ser una introducción al tema, por lo cual hay aspectos que son obviados. Aun así, se invita al lector a profundizar en la materia.
</p>

## Distribución Bidimensional de Esfuerzos

### Medio Elástico Lineal

<p style='text-align: justify;'>El ingeniero alemán Ernst Gustav Kirsch fue el primero en establecer, en 1898, una solución sobre la distribución de esfuerzos en dos dimensiones de una excavación en un medio sólido linealmente elástico, para una sección circular (Hoek & Brown, 1980a).
Las ecuaciones de Kirsch son una solución analítica en coordenadas polares del campo de esfuerzos inducidos alrededor de una excavación subterránea circular, alejada de otras excavaciones, y sometida a un campo de esfuerzos biaxial. Ver Figura 1.1.</p>

<p style='text-align: justify;'> Las ecuaciones 1.1 se utilizan para calcular los esfuerzos inducidos en un punto del macizo rocoso descrito por las coordenadas (r; $\theta$) alrededor de una galería de radio "a". Estas ecuaciones surgen de una función de Airy en coordenadas polares. </p>

\begin{equation}
\sigma_{\theta\theta} = \sigma_0 \left(\frac{1+k}{2} \right) \left(1+\frac{a^2}{r^2} \right) + \sigma_0 \left(\frac{1-k}{2} \right) \left(1+ \frac{3a^4}{r^4}\right) \cos(2 \theta)
\end{equation} 


\begin{equation}
\sigma_{rr} = \sigma_0 \left(\frac{1+k}{2} \right) \left(1-\frac{a^2}{r^2} \right) - \sigma_0 \left(\frac{1-k}{2} \right) \left(1- \frac{4a^2}{r^2} + \frac{3a^4}{r^4} \right) \cos(2 \theta)
\end{equation}
<div style="text-align: right"> (1.1) </div>
\begin{equation}
\sigma_{r\theta} = \sigma_0 \left( \frac{1-k}{2} \right) \left(1+ \frac{2a^2}{r^2} - \frac{3a^4}{r^4} \right) \sin(2 \theta)
\end{equation}


    
#### <center> Figura  1.1 : Esquema del modelo de las ecuaciones de Kirsch en dos dimensiones. </center>


En las ecuaciones anteriores, $\sigma_0$ corresponde al esfuerzo vertical in situ, generalmente determinado como la presión litostática. Es descrito por la ecuación 1.2.

\begin{equation}
\sigma_0 = \rho gz           
\end{equation}
<div style="text-align: right"> (1.2) </div>

Donde $\rho$ es la densidad de la roca, $g$ es la aceleración de gravedad (9.8 $\frac{m}{s^2}$) y $z$ es la profundidad a la cual se va a construir el túnel.
El esfuerzo horizontal corresponde al término k$\sigma_0$, donde k es una constante, y se define como la razón entre el esfuerzo horizontal y vertical. Su valor suele variar entre 0.5 y 3.

Los desplazamientos producidos pueden ser calculados mediante las ecuaciones 1.3.

</p>
\begin{equation}
u_r = - \frac{\sigma_0 a^2}{4 G r} \left[ 1+k-(1-k) \left[4(1- \nu) - \frac{a^2}{r^2} \right ]\cos{(2\theta)} \right]
\end{equation}
<div style="text-align: right"> (1.3) </div>
\begin{equation}
u_\theta = - \frac{\sigma_0 a^2}{4 G r} (1-k)\left[ 2(1-2\nu) + \frac{a^2}{r^2} \right] \sin{(2\theta)}
\end{equation}

Las Figuras 1.2 y 1.3 muestran la distribución de esfuerzos $\sigma_{rr}$, $\sigma_{\theta\theta}$ y $\sigma_{r\theta}$ condiciones tectónicas distensivas (k < 1) y compresivas (k > 1), respectivamente.
    


#### <center> Figura  1.2 : Esfuerzos inducidos  $\sigma_{rr}$ , $\sigma_{\theta\theta}$ y $\sigma_{r\theta}$ en una galería con k = 0.5 </center>


#### <center> Figura  1.3 : Esfuerzos inducidos  $\sigma_{rr}$ , $\sigma_{\theta\theta}$ y $\sigma_{r\theta}$ en una galería con k = 1.5 </center>


<p style='text-align: justify;'>A partir de las ecuaciones 1.1 es posible determinar los esfuerzos principales $\sigma_1$ y $\sigma_3$ resolviendo los valores de $\lambda$ que satisfacen la ecuación 1.4, la cual surge de la ecuación $ det (\sigma-\lambda I)=0$ al escribir el tensor de esfuerzos $\sigma$ en su forma polar. </p>

\begin{equation}
\begin{vmatrix}
\sigma_{rr}-\lambda & \sigma_{r \theta} \\
\sigma_{r \theta} & \sigma_{\theta \theta}-\lambda \\
\end{vmatrix} =0
\end{equation}
<div style="text-align: right"> (1.4) </div>

<p style='text-align: justify;'>Las Figuras 1.4 y 1.5 muestran los esfuerzos inducidos $\sigma_1$ y $\sigma_3$ alrededor de una galería. Como se aprecia en la Figura 1.15, cuando existe un régimen de esfuerzos distensivo, es decir, el esfuerzo vertical es mayor que el esfuerzo horizontal, los esfuerzos inducidos se concentran en las paredes de la galería. Por el contrario, cuando existen condiciones tectónicas compresivas, es decir, el esfuerzo horizontal es mayor que el esfuerzo vertical, los esfuerzos se concentran en el techo, como puede verse en la Figura 1.5.

### Ejemplo aplicación

Para la siguiente excavacion obtenga los graficos de distribucion de esfuerzos de $\sigma_{rr}$, $\sigma_{\theta\theta}$ y $\sigma_{r\theta}$




```python

#codigo para graficar los esfuerzos en su forma polar
import numpy as np 
import matplotlib.pyplot as plt 

n= 1000
Lim=10
a=2
s0=15
r= np.linspace(a,Lim*(2**2), n)
o=np.linspace(0, 2 * np.pi, n)
r,o= np.meshgrid(r,o)
x = r*np.cos(o)
y =  r*np.sin(o)

k=1.5
Lim=10

Soo= s0*((1+k)/2)*(1+(a**2)/(r**2))+s0*((1-k)/2)*(1+(3*(a**4)/(r**4)))*np.cos(2*o)
Srr= s0*((1+k)/2)*(1-(a**2)/(r**2))-s0*((1-k)/2)*(1-4*(a**2)/(r**2)+3*(a**4)/(r**4))*np.cos(2*o)
Sro= s0*((1-k)/2)*(1+2*(a**2)/(r**2)- 3*(a**4)/(r**4))*np.sin(2*o)
S1=22.5+(8*np.cos(2*o)/(r**(2)))+((90/(r**(2)))+(7.5*(1-8/(r**(2))+48/(r**(4)))*np.cos(2*o))**(2)+(56.25*np.sin(2*o)**(2))*(1+8/(r**(2))-48/(r**(4)))**(2))**(0.5)
S3= 22.5+(8*np.cos(2*o)/r**(2))-((90/r**(2))+(7.5*(1-8/r**(2)+48/r**(4))*np.cos(2*o))**(2)+(56.25*np.sin(2*o)**(2))*(1+8/r**(2)-48/r**(4))**(2))**(0.5)

fig, ax = plt.subplots(figsize=[5, 4])
ax.set(aspect=1, xlim=(-Lim,Lim), ylim=(-Lim,Lim))
plt.contourf(x, y, Soo,cmap='jet') 
ax.set_title('$\sigma_{oo}$', loc = "center", fontdict = {'fontsize':14})
cbar=plt.colorbar() 
cbar.set_label('MPa') 
plt.show() 

fig, ax = plt.subplots(figsize=[5, 4])
ax.set(aspect=1, xlim=(-Lim, Lim), ylim=(-Lim, Lim))
plt.contourf(x, y, Srr,cmap='jet') 
ax.set_title('$\sigma_{rr}$', loc = "center", fontdict = {'fontsize':14})
cbar=plt.colorbar() 
cbar.set_label('MPa')
plt.show() 

fig, ax = plt.subplots(figsize=[5, 4])
ax.set(aspect=1, xlim=(-Lim, Lim), ylim=(-Lim, Lim))
plt.contourf(x, y, Sro,cmap='jet') 
ax.set_title('$\sigma_{ro}$', loc = "center", fontdict = {'fontsize':14})
cbar=plt.colorbar() 
cbar.set_label('MPa')
plt.show()
```
