## Esfuerzos

<p style='text-align: justify;'> El concepto de esfuerzo o stress es utilizado para describir la intensidad de las fuerzas internas inducidas en un cuerpo bajo la influencia de un conjunto de fuerzas aplicadas en su superficie. El estado de esfuerzo en un punto de un sólido se define en términos de fuerza por unidad de área actuando en las superficies ortogonalmente orientadas de un cuerpo libre infinitesimal centrado en el punto.

<p style='text-align: justify;'>Si se utiliza un sistema de referencia cartesiano, el elemento infinitesimal es un cubo cuyas caras están orientadas con los ejes coordenados.</p>

<p style='text-align: justify;'>En dos dimensiones el tensor de esfuerzos queda definido por la matriz siguiente:
</p>

\begin{equation}
\sigma=
\begin{bmatrix}
\sigma_{xx} & \sigma_{xy} \\
\sigma_{yx} & \sigma_{yy} \\
\end{bmatrix}
\end{equation}
<div style="text-align: right"> (1.1) </div>

<p style='text-align: justify;'> En este contexto, los esfuerzos normales se escriben como $\sigma_{xx}$ y  $\sigma_{yy}$ y los esfuerzos cortantes como $\sigma_{xy}$ y  $\sigma_{yx}$.</p>

<p style='text-align: justify;'>Puede demostrarse que, para que un sólido se encuentre en equilibrio, la matriz anterior debe ser necesariamente simétrica, lo que implica que las cuatro componentes de esfuerzos contenidas en la matriz se reducen a tres, ya que se tiene la condición $\sigma_{xy}$ = $\sigma_{yx}$.</p>

<p style='text-align: justify;'>En tres dimensiones, el tensor de esfuerzos está determinado por seis componentes
independientes: </p>

\begin{equation}
\sigma=
\begin{bmatrix}
\sigma_{xx} & \sigma_{xy} & \sigma_{xz}\\
\sigma_{yx} & \sigma_{yy} & \sigma_{yz}\\
\sigma_{zx} & \sigma_{zy} & \sigma_{zz}
\end{bmatrix}
\end{equation}
<div style="text-align: right"> (1.2) </div>

<p style='text-align: justify;'> Donde se debe cumplir $\sigma_{xy}$ = $\sigma_{yx}$, $\sigma_{xz}$ = $\sigma_{zx}$ y $\sigma_{yz}$ = $\sigma_{zy}$ para mantener en equilibrio estático el cuerpo.
Ya sea un cuerpo en dos o tres dimensiones, si se realiza un corte en este, se requiere la acción de una fuerza $\Delta$ F para mantenerlo en equilibrio. Tal como se muestra en la Figura, esta fuerza puede ser descompuesta en dos vectores, uno normal
($\Delta$N ) y otro paralelo ($\Delta$T) a la superficie. </p>


#### <center> Figura  1.1: Fuerzas internas que mantienen al sólido en equilibrio. </center>


<p style='text-align: justify;'>Estas fuerzas pueden ser transformadas a esfuerzos normales y tangenciales al dividirlas  por  un  area  ∆A  lo  suficientemente  pequenña.  Matemáticamente  esto  se puede expresar mediante las ecuaciones 1.3 y 1.4. </p>

\begin{equation}
\sigma_{n}= \lim_{\Delta A \to \infty}\frac{\Delta N}{\Delta A}
\end{equation}
<div style="text-align: right"> (1.3) </div>
\begin{equation}
\sigma_{t}= \lim_{\Delta A \to \infty}\frac{\Delta T}{\Delta A}
\end{equation}
<div style="text-align: right"> (1.4) </div>


<p style='text-align: justify;'>Conocido el tensor de esfuerzos σ y el vector normal a un plano $\vec{n}$ es posible calcular el esfuerzo normal mediante la ecuación.</p>
<p style='text-align: justify;'>El término $\sigma$ $\vec{n}$ da como resultado un vector que suele denominarse vector de tracciones ($\vec{t}$).</p>


\begin{equation}
\sigma_n = \sigma \vec{n} \cdot \vec{n}
\end{equation}
<div style="text-align: right"> (1.5) </div>

### Esfuerzo Plano
<p style='text-align: justify;'>Se entiende por estado de esfuerzo plano cuando los esfuerzos solo ocurren en el plano xy, por lo que $\sigma_{zz}$ = 0, implicando que $\sigma_{ij}$ solo se encuentra en función de las coordenadas x e y. Para utilizar este estado se debe garantizar que el cuerpo es lo suficientemente delgado en la dimensión z. Se utiliza para modelar los esfuerzos en placas delgadas.</p>

<p style='text-align: justify;'>En un sistema de coordenadas cartesianas de dos dimensiones el problema puede ser analizado mediante un elemento infinitesimal rectangular que pertenece a un sólido y en el cual se encuentra actuando una condición de esfuerzos planos. Si se realiza un corte imaginario en un plano que forma un ángulo $\theta$ con la vertical, tal como se puede observar en la Figura , al efectuar un análisis de equilibrio de fuerzas y momentos es posible deducir las  ecuaciones  para determinar los esfuerzos tangenciales ($\sigma_{t}$) y normales ($\sigma_{n}$) actuando sobre dicho plano.</p>



#### <center> Figura  1.3: Esfuerzos actuando sobre un elemento infinitesimal en dos dimensiones. </center>

\begin{equation}
\sigma_{n} = \frac {\sigma_{xx} + \sigma_{yy}}{2} + (\frac {\sigma_{xx} - \sigma_{yy}}{2}) \cos{2\theta} + \sigma_{xy} \sin {2\theta}
\end{equation}
<div style="text-align: right"> (1.6) </div>

\begin{equation}
\sigma_{t} = \frac {\sigma_{xx} + \sigma_{yy}}{2}  \sin {2\theta} +  \sigma_{xy} \cos {2\theta}
\end{equation}

```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


#sxx = int(input("ingrese el valor de sigma xx en MPa"))
#syy = int(input("ingrese el valor de sigma yy en MPa"))
#sxy = int(input("ingrese el valor de sigma xy en MPa"))

sxx = 10 #MPa
syy = 8  #MPa
sxy = 5  #MPa

#Rango para iterar el grafico
at = np.linspace(0, 2.5*np.pi, 256, endpoint=True)  #at=angulo theta

#ecuaciones
sn = ((sxx+syy)/2)+ ((sxx-syy)/2) *(np.cos(2*at)) + sxy * np.sin (2*at)

st = ((sxx-syy)/2)*np.sin(2*at)+ sxy*np.cos(2*at)

#graficas
plt.plot(at, sn, color="red",linewidth=0.5, linestyle="-") #grafico de sigma normal, color,grosor y estilo de la linea de la curva
plt.plot(at, st, color="blue",linewidth=0.5, linestyle="-") #grafico sigma tangencial,color,grosor y estilo de la linea de la curva

plt.xlabel(r'Ángulo $\theta$')  # titulo eje x
plt.ylabel('Esfuerzo [MPa]') # titulo eje y

plt.grid(color='black', linestyle='dotted', linewidth=0.5) #grilla y color,tipo y grosor de la linea de grilla 



plt.show()

print("Como puede observarse, cuando el esfuerzo normal alcanza su valor máximo o mínimo, el esfuerzo tangencial es cero. Esto ocurre tanto en problemas de dos como de tres dimensiones.")
```


## Rotación de Esfuerzos

El estado de esfuerzos en el plano xy puede transformarse a un nuevo sistema de referencia x'y' realizando una rotación de un ángulo $\theta$, esquematizada en la siguiente figura 1.4.


#### <center> Figura  1.4: Rotación del estado de esfuerzos. </center>

El nuevo estado de esfuerzos, constituido por $\sigma_{x'x'}$ , $\sigma_{y'y'}$ y $\sigma_{x'y'}$ puede ser determinado mediante las ecuaciones 1.7.

\begin{equation}
\sigma_{x'x'} = \sigma_{x'}= \frac{\sigma_{x}+\sigma_{y}}{2} + \frac{\sigma_{x}-\sigma_{y}}{2}\cos(2 \theta) + \sigma_{xy}\sin(2 \theta)
\end{equation}
\begin{equation} 
\sigma_{y'y'} = \sigma_{y'}= \frac{\sigma_{x}+\sigma_{y}}{2} - \frac{\sigma_{x}-\sigma_{y}}{2}cos(2 \theta) - \sigma_{xy}sin(2 \theta)
\end{equation}
<div style="text-align: right"> (1.7) </div>
\begin{equation} 
\sigma_{x'y'} = -\frac{\sigma_{x} - \sigma_{y}}{2}sin(2 \theta) + \sigma_{xy}cos(2\theta)
\end{equation}

Una rotación del sistema de referencia en tres dimensiones suele realizarse mediante cálculo matricial. El nuevo estado de esfuerzo $\sigma'$ con base en un sistema de referencia $x'y'z'$ se obtiene mediante la ecuación 1.8.

\begin{equation} 
\sigma' = R\sigma R^T
\end{equation}
<div style="text-align: right"> (1.8) </div>

Donde $R$ es la matriz de rotación y $R^T$ es su matriz transpuesta. Cada fila de la matriz $R$ está formada por los vectores normales unitarios $l$, $m$ y $n$ que forman el nuevo sistema de referencia $x'$, $y'$ y $z'$.
El análisis anterior puede también desarrollarse para un caso bidimensional, obteniéndose expresiones análogas a las ecuaciones 1.7.

## Esfuerzos Principales
En cualquier punto sometido a esfuerzos es posible encontrar tres planos, ortogonales entre sí, en los que los esfuerzos tangenciales son nulos; estos planos se denominan planos principales de esfuerzo. Los esfuerzos normales que actúan sobre ellos se denominan esfuerzos principales ($ \sigma_1 , \sigma_2 y  \sigma_3$). Para determinar los esfuerzos principales y las direcciones principales en dos dimensiones se puede recurrir al uso de derivadas para determinar máximos y mínimos de funciones. Sin embargo, esto en tres dimensiones no es posible y se suele utilizar cálculo matricial para ello. Los esfuerzos principales pueden ser determinados al resolver la siguiente expresión:

\begin{equation} 
det (\sigma-\lambda I)=0
\end{equation}
<div style="text-align: right"> (1.9) </div>

Donde $\sigma$ es el tensor de esfuerzos, $\lambda$ es una constante e $I$ es la matriz identidad.
Al resolver los valores de $\lambda$ que satisfacen esta expresión se determinan los tres esfuerzos principales.
La dirección principal $r_i$ asociada al esfuerzo principal $\sigma_i$ (con i = 1; 2; 3) está dada por la ecuación:

\begin{equation} 
\vec{r_i} = \frac{1}{\sqrt{a^2+b^2+c^2}}(a\hat{\imath}+b\hat{\jmath}+c\hat{k})
\end{equation}
<div style="text-align: right"> (1.10) </div>

Donde las constantes a, b y c son determinadas con las siguientes expresiones:

\begin{equation}
a = 
\begin{vmatrix}
\sigma_{yy}-\sigma_i & \sigma_{yz} \\
\sigma_{zy} & \sigma_{zz}-\sigma_i \\
\end{vmatrix}
\end{equation}

<div style="text-align: right"> (1.11) </div>

\begin{equation}
b = -
\begin{vmatrix}
\sigma_{yx} & \sigma_{yz} \\
\sigma_{zx} & \sigma_{zz}-\sigma_i \\
\end{vmatrix}
\end{equation}


\begin{equation}
c = 
\begin{vmatrix}
\sigma_{yx} & \sigma_{yy}-\sigma_i \\
\sigma_{zx} & \sigma_{zy} \\
\end{vmatrix}
\end{equation}

## Deformaciones
<p style='text-align: justify;'>La aplicación de fuerzas o variaciones de temperatura generarán cambios en la posición relativa de los puntos dentro de un sólido rígido, es decir, este cambiará su forma. Estas deformaciones o strain pueden ser cuantificadas conociendo la distribución de los desplazamientos de todos los puntos del sólido. La deformación es una variable adimensional.</p>

### Deformaciones Unidimensionales
<p style='text-align: justify;'>Una deformación unidimensional, como por ejemplo la deformación de una barra de sección delgada, puede ser determinada como el cambio de longitud de la barra ($\Delta l$) debido a la acción de una fuerza externa en uno de sus extremos con respecto a su longitud inicial ($l$).</p>



\begin{equation} 
\varepsilon= \frac{\Delta l}{l}
\end{equation}
<div style="text-align: right"> (1.12) </div>

<p style='text-align: justify;'>Por otra parte, al presentarse un cambio de temperatura en el sólido, este también experimentará una deformación axial, denominada deformación térmica.</p>

\begin{equation} 
\varepsilon_T= \alpha \Delta l T
\end{equation}
<div style="text-align: right"> (1.13) </div>

<p style='text-align: justify;'>Donde $\Delta T $ es el cambio de temperatura y $\alpha$ es el coeficiente de dilatación térmica, cuyo valor, según la clase de roca, puede variar entre 4 $\times  10^{-6}$ y 12 $\times 10^-6 $ $[K^{-1}]$.</p>


### Deformación Plana y Rotación de Deformaciones

<p style='text-align: justify;'>Se habla de deformación plana cuando las deformaciones solo ocurren en el plano xy, por lo que la deformación en el eje z es nula. Para utilizar este estado se debe garantizar que la geometría no cambia a lo largo de la dimensión z. Un ejemplo en geotecnia sería un túnel o un talud lo suficientemente largo. Sin embargo, los túneles o los taludes no son infinitos. Se ha demostrado que la solución matemática es adecuada cuando la geometría se mantiene constante durante al menos cinco veces las dimensiones en el plano xy (Rocscience, 2019).</p>
<p style='text-align: justify;'>Un elemento rectangular dentro de un sólido puede verse afectado por tres tipos distintos de deformaciones (Figura 1.6). El primer tipo de deformación admisible es la componente de deformación normal o extensional en la dirección n (en dos dimensiones esta puede ser en la dirección x o en la dirección y), la cual se define como el cambio de longitud con respecto a la longitud original de las fibras del elemento orientadas en esa dirección. El segundo tipo de deformación admisible es la deformación de cizalle o al corte, la cual involucra cambios en los ángulos de la geometría del sólido. La deformación de cizalle es definida como el cambio en el ángulo entre dos direcciones originalmente ortogonales en el material.</p>



#### <center> Figura  1.5: Deformaciones características de un elemento rectangular. </center>

<p style='text-align: justify;'>Un sólido bidimensional puede estar sometido a una combinación de los distintos tipos de deformaciones mostrados en la Figura 1.5. El tensor de deformación " suele formularse mediante las expresiones 1.14 y 1.15. Ambos tensores son iguales, únicamente difieren en la notación de la deformación de cizalle.</p>

\begin{equation}
\varepsilon = 
\begin{bmatrix}
\varepsilon_{xx} & \varepsilon_{xy} \\
\varepsilon_{yx} & \varepsilon_{yy} \\
\end{bmatrix}
\end{equation}
<div style="text-align: right"> (1.14) </div>
\begin{equation}
\varepsilon = 
\begin{bmatrix}
\varepsilon_{xx} & \frac{1}{2} \gamma_{xy} \\
\frac{1}{2} \gamma_{xy} & \varepsilon_{yy} \\
\end{bmatrix}
\end{equation}
<div style="text-align: right"> (1.15) </div>

<p style='text-align: justify;'>Para que las deformaciones sean físicamente posibles dentro de un sólido se debe cumplir que $\varepsilon_{xy} = \varepsilon_{yx}$ o, equivalentemente, que $\gamma_{xy} = \gamma_{yx}$.</p>

<p style='text-align: justify;'>Al igual que con los tensores de esfuerzo, es posible transformar un tensor de deformación a partir de un sistema de referencia xy a un nuevo sistema de referencia $x'y'$ , rotado un ángulo $\theta$ con respecto al sistema inicial, tal como se muestra en la Figura 1.6. Las nuevas componentes de deformación $\varepsilon_{x'}$ ,  $\varepsilon_{y'}$ y   $\varepsilon_{x'y'}$ pueden ser calculadas mediante las ecuaciones 1.16.</p>


#### <center> Figura  1.6: Rotación en dos dimensiones. </center>

\begin{equation}
\varepsilon_{x'}= \frac{\varepsilon_{x}+\varepsilon_{y}}{2} + \frac{\varepsilon_{x}-\varepsilon_{y}}{2}\cos(2 \theta) + \frac{\gamma_{xy}}{2}\sin(2 \theta)
\end{equation}
\begin{equation} 
\varepsilon_{y'}= \frac{\varepsilon_{x}+\varepsilon_{y}}{2} - \frac{\varepsilon_{x}-\varepsilon_{y}}{2}cos(2 \theta) - \frac{\gamma_{xy}}{2}sin(2 \theta)
\end{equation}
<div style="text-align: right"> (1.16) </div>
\begin{equation} 
\gamma_{x'y'} = -\frac{\varepsilon_{x} - \varepsilon_{y}}{2}sin(2 \theta) + \frac{\gamma_{xy}}{2} cos(2\theta)
\end{equation}

### Deformaciones Principales

<p style='text-align: justify;'>Dentro de un sólido sometido de deformaciones es posible encontrar tres ejes ortogonales entre sí, en los cuales las deformaciones de cizalle son nulas. Estos ejes se denominan direcciones principales de deformación, y las deformaciones normales que en ellos actúan se designan como deformaciones principales ($\varepsilon_1$,$\varepsilon_2$ y $\varepsilon_3$). Las deformaciones principales y sus direcciones principales pueden obtenerse mediante el cálculo de valores y vectores propios utilizando las ecuaciones 1.9 y 1.10, respectivamente, sustituyendo el tensor de esfuerzos $\sigma$ por el tensor de deformación $\varepsilon$ .</p>

### Rosetas de Deformación

<p style='text-align: justify;'>La deformación unitaria normal en un sólido sometido a carga se mide utilizando un *strain gauge* o extensiómetro de resistencia eléctrica, el cual consiste en una red de alambres, o una pieza de hoja metálica adherida a la superficie del espécimen. Con frecuencia se determinan las deformaciones unitarias normales en un punto de la superficie del sólido mediante un conjunto de tres *strain gauges* dispuestos de una forma determinada (Figura 1.7). A esta disposición se le denomina roseta de deformación. Cada configuración se diferencia por el ángulo de separación de los *strain gauges*. Las rosetas más comunes corresponden a extensiómetros dispuestos con una separación de 45 $^{\circ}$ o 60  $^{\circ}$ entre ellos.</p>



#### <center> Figura  1.7: Roseta de Deformación. </center>

<p style='text-align: justify;'> Una vez que se determinan las lecturas de deformación en los tres extensiómetros ($\varepsilon_a$,$\varepsilon_b$ y $\varepsilon_c$), estas pueden emplearse para determinar el estado de deformación en el punto utilizando las ecuaciones 1.16, resultando el sistema de ecuaciones 1.17 con tres ecuaciones y tres incógnitas ($\varepsilon_{xx}$,$\varepsilon_{yy}$ y $\varepsilon_{xy}$).</p>

\begin{equation}
\varepsilon_{a}= \frac{\varepsilon_{xx}+\varepsilon_{yy}}{2} + \frac{\varepsilon_{xx}-\varepsilon_{yy}}{2}\cos(2 \theta_a) + \frac{\gamma_{xy}}{2}\sin(2 \theta_a)
\end{equation}

\begin{equation}
\varepsilon_{b}= \frac{\varepsilon_{xx}+\varepsilon_{yy}}{2} + \frac{\varepsilon_{xx}-\varepsilon_{yy}}{2}\cos(2 \theta_b) + \frac{\gamma_{xy}}{2}\sin(2 \theta_b)
\end{equation}
<div style="text-align: right"> (1.17) </div>
\begin{equation}
\varepsilon_{c}= \frac{\varepsilon_{xx}+\varepsilon_{yy}}{2} + \frac{\varepsilon_{xx}-\varepsilon_{yy}}{2}\cos(2 \theta_c) + \frac{\gamma_{xy}}{2}\sin(2 \theta_c)
\end{equation}
