# Guía simple de Geomecánica
<p style='text-align: justify;'> Los macizos rocosos no están formados solo por un gran bloque de roca intacta. Por lo general, se encuentran en bloques separados por los conjuntos de discontinuidades que lo forman. Estas uniones cambian el comportamiento de la roca ante las tensiones. Es por eso que a Hoek se le ocurrió el conocimiento de las fallas por fragilidad y a Brown el conocimiento del comportamiento del macizo rocoso con sus discontinuidades y juntos, identificaron una relación empírica de las descripciones del campo del macizo rocoso (Clasificación del macizo rocoso) con un criterio de falla para los macizos rocosos. Su investigación, junto con la investigación de otros autores (Palmstrom, Bieniawski, Barton, etc.), son un gran aporte a la geomecánica y estudio del comportamiento de los macizos rocosos.
Al aplicar las ecuaciones derivadas de sus resultados, los geomecánicos pueden estimar bajo qué tensiones una masa rocosa entrará en estado de falla con solo observaciones de campo escrupulosas, mediciones precisas y una lectura atenta de sus artículos.

## Calcular criterios de fractura de Hoek-Brown con Python

<p style='text-align: justify;'> El criterio de fractura de Hoek-Brown (Hoek & Brown, 1980a; Hoek & Brown, 1980b) fue desarrollado con el objeto de entregar una línea base para el diseño de excavaciones subterráneas. Durante la época de publicación del criterio original no existían métodos apropiados para la estimación de la resistencia de macizos rocosos. Es por ello que la meta fue formular una ecuación que pudiese ser ligada a información geológica obtenible en terreno (Hoek & Marinos, 2007). La ecuación 1.1 expresa el criterio en su forma original.</p>

\begin{equation}
\sigma_{1} = \sigma_{3} + \sigma_{ci} \left(m \frac{\sigma_{3}}{\sigma_{ci}}+ s \right) ^{0.5}
\end{equation}
<div style="text-align: right"> (1.1) </div>

<p style='text-align: justify;'> Donde $\sigma_{1}$ y $\sigma_{3}$ son los esfuerzos principales efectivos mayor y menor al momento de la fractura, respectivamente, $\sigma_{ci}$ es la resistencia a la compresión uniaxial de la roca intacta, m y s son constantes del material, donde s=1 es para roca intacta. Originalmente se utilizó el RMR (Bieniawski, 1973; Bieniawski, 1976; Bieniawski, 1989) para relacionar el criterio a las observaciones geológicas de campo, debido a que ya se encontraba ampliamente difundido y contaba con buena aceptación.
Sin embargo, con el paso del tiempo fue haciéndose notable la dificultad de aplicación del RMR en macizos rocosos de mala calidad, debido a que estos difieren significativamente de un macizo de roca dura con bloques firmemente trabados, sobre el cual el criterio original fue formulado, por lo que se decidió actualizar el criterio (ecuaciones 1.2) introduciendo el GSI (Hoek et al., 1995) en reemplazo del RMR.
    
</p>

\begin{equation}
\sigma_{1} = \sigma_{3} + \sigma_{ci}\left (m \frac{\sigma_{3}}{\sigma_{ci}}+ s \right) ^{a}
\end{equation}

Para GSI > 25:
\begin{equation}
\frac{m_b}{m_i} = exp \left(\frac{GSI - 100}{28} \right)
\end{equation}

\begin{equation}
s = exp \left(\frac{GSI - 100}{9} \right)
\end{equation}
\begin{equation}
a= 0.5
\end{equation}

Para GSI < 25:
\begin{equation}
s=0
\end{equation}
\begin{equation}
a=0.65 - \frac{GSI}{200}
\end{equation}
<div style="text-align: right"> (1.2) </div>

<p style='text-align: justify;'> Donde, $\sigma_{1}$ y $\sigma_{3}$ son los esfuerzos principales efectivos mayor y menor al momento de la fractura, respectivamente, $\sigma_{ci}$ es la resistencia a la compresión uniaxial de la roca intacta, mb es el valor de la constante m para el macizo rocoso, s y a son constantes que dependen de las características del macizo rocoso, y mi es una constante de material para roca intacta.
Hoek et al. (2002) realizaron una revisión importante al criterio e incorporando nuevas relaciones entre los parámetros m, s, a y GSI. Además, se introdujo una nueva variable denominada factor de perturbación (D), ligada al grado de alteración al que el macizo rocoso ha sido sometido debido a tronaduras y relajación de esfuerzos. </p>


El criterio generalizado de Hoek-Brown se expresa como (ecuaciones 1.3):

\begin{equation}
\sigma´_{1} = \sigma´_{3} + \sigma_{ci} \left(m \frac{\sigma´_{3}}{\sigma_{ci}}+ s \right) ^{a}
\end{equation}

\begin{equation}
m_b= m_i  exp \left(\frac{GSI - 100}{28-14D}\right)
\end{equation}

\begin{equation}
s = exp \left (\frac{GSI - 100}{9-3D} \right)
\end{equation}
<div style="text-align: right"> (1.3) </div>
\begin{equation}
a=\frac{1}{2} + \frac{1}{6} \left( e^{\frac{-GSI}{15}} -  e^{\frac{-20}{3}}\right)
\end{equation}

<p style='text-align: justify;'> La Figura 1.1 ilustra el comportamiento no lineal de la envolvente de fractura del criterio.
La constante de roca mi es obtenida a través de análisis estadístico de ensayos de compresión triaxial. Si esto no es posible de realizar, puede ser estimada como $\sigma_{ci}$=$\sigma_t$ (Brown, 2008). La Tabla 1.2 entrega valores de mi para diferentes tipos de roca intacta.
El factor de perturbación D depende del grado de alteración del macizo rocoso según las actividades de tronadura y/o relajación de esfuerzos. Su valor puede ir desde 0 para macizos rocosos inalterados, hasta 1 para macizos rocosos fuertemente alterados. Para su selección puede consultarse la Tabla 1.3. </p>
<p style='text-align: justify;'>  La resistencia a la compresion uniaxial del macizo rocoso puede obtenerse al aplicar
$\sigma'_3$ = 0, resultando en la ecuación 1.4. </p>

\begin{equation}
\sigma_{cm} = \sigma_{ci} s^{a}
\end{equation}
<div style="text-align: right"> (1.4) </div>

Procediendo de forma análoga, se desprende la resistencia tensional del macizo rocoso (ecuación 1.5): 

\begin{equation}
\sigma_{tm} = - {\frac{\sigma_{ci} s}{m_b}}
\end{equation}
<div style="text-align: right"> (1.5) </div>

<p style='text-align: justify;'> La ecuación 1.5 se deduce al considerar $\sigma'_1 = \sigma'_3 = \sigma_{tm}$. Esto describe una situación biaxial de esfuerzos. Hoek (1983) postuló que, para materiales frágiles, la resistencia tensional uniaxial es igual a la biaxial (Hoek et al., 2002). </p>

### Graficando el Criterio de Hoek & Brown

Comencemos con un ejemplo de descripción de material rocoso: 
<p style='text-align: justify;'>
Roca afanítica extrusiva de composición intermedia. Degradado con una fina capa de alteración. Se requirió un golpe firme de un martillo geológico para fracturar una muestra. 
Peso unitario = 0.0279 MN / m³
Y una descripción de masa rocosa de ejemplo para la misma pendiente:
Pendiente de roca en una zona minera a cielo abierto con gran perturbación.
El macizo rocoso es de bloques. Tres conjuntos de articulaciones identificadas. Dos son lisos y planos, uno es irregular y ondulado. Los espacios para los tres conjuntos de juntas son de aproximadamente 250 mm, 350 mm y 600 mm.</p>

Usemos el peor de los casos para la resistencia a la compresión uniaxial (σci) y la constante del material (mi).
Se fracturó con un golpe firme, por lo que podemos ponerlo en R3: la resistencia a la compresión uniaxial es de 25 MPa.
Además, la constante del material para la andesita es 20.
El factor de perturbación es 1.0 para la minería a cielo abierto con voladuras elevadas.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-dark')

#Ejemplo GSI

Ja  =  3.0 
Jr  =  1.0

Jv  =  ( 1/0.25 ) + ( 1 /0.35 ) + ( 1 /0.60 ) 
print("Jv = " ,round(Jv,2))

RQD  =  110 - 2.5 * Jv 
print("RQD = " ,round(RQD,2))

GSI  =  (( 52  *  ( Jr / Ja ))  /  ( 1  +  ( Jr / Ja )))  +  ( RQD  /  2 ) 
print("GSI = ",round(GSI,2))


#variables ingresables
sigci = 25.0
mi = 20.0
D = 1.0
unit_weigth = 0.0279
depth = 25

mb = mi * np.exp((GSI-100)/(28-(14*D)))
s = np.exp((GSI-100)/(9-(3*D)))
a = (1/2)+(1/6)*(np.exp(-GSI/15)-np.exp(-20/3))

#ecuaciones Hoek & Brown
def hoek_brown(min_sig3, max_sig3, num_values):
    sig3 = np.linspace(min_sig3, max_sig3, num_values, endpoint=True)
    sig1 = sig3 + sigci * ((mb * (sig3/sigci))+s)**a
    sig_array = pd.DataFrame(data=np.vstack([sig3, sig1]).T, columns=['sig3', 'sig1'])

    sig_array['ds1ds3'] = 1+a*mb*(mb*(sig3/sigci)+s)**(a-1)
    sig_array['sign'] = ((sig1 + sig3)/2) - ((sig1 - sig3)/2) * (sig_array['ds1ds3'] - 1) / (sig_array['ds1ds3'] + 1)
    sig_array['tau'] = (sig1-sig3)*np.sqrt(sig_array['ds1ds3'])/(sig_array['ds1ds3']+1)

    return sig_array
#Veamos cómo va usando 15 valores de 0 a 1.
sig_array = hoek_brown(0, 1, 15)
sig_array.dropna(inplace=True)
sig_array

# Graficos 
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(121)

sig1_formula = r'\sigma_1^{´}=\sigma_3^{´} + \sigma_{ci} \times (m_b \times \frac{\sigma_3^{´}}{\sigma_{ci}} + s)^a'

ax.plot(sig_array.sig3.values, sig_array.sig1.values, 'bo-',
        label='Hoek-Brown: ' + sig1_formula.join(r'$$')); 
ax.set_xlabel(u'Esfuerzo Principal, $\sigma_3$, MPa', fontsize=12)
ax.set_ylabel(u'Esfuerzo Principal, $\sigma_1$, MPa', fontsize=12)

ax.grid(); ax.legend(fontsize='x-large')

ax = fig.add_subplot(122)

tau_formula = r'\tau^{´}=(\sigma_1^{´}-\sigma_3^{´}) \frac {\sqrt{d\sigma_1^{´}/d\sigma_3^{´}}} {d\sigma_1^{´}/d\sigma_3^{´} + 1}'

ax.plot(sig_array['sign'], sig_array['tau'], 'bo-')
ax.set_xlabel(u'Esfuerzo Normal, $\sigma_n$, MPa', fontsize=12)
ax.set_ylabel(u'Esfuerzo Cortante, $\u03C4$, MPa', fontsize=12)

# Add Mohr circles for the invervals that were defined above.

centers = ((sig_array.sig1.values - sig_array.sig3.values) / 2) + sig_array.sig3.values
radius = sig_array.sig1.values - centers

for r, c in zip(radius, centers):
    ax.add_patch(plt.Circle([c, 0], r, facecolor='none', ec='k', alpha=.2))

ax.grid()
ax.set_aspect('equal')


fig.suptitle('Criterio fractura de Hoek & Brown ', fontsize=20, y=1.05)
fig.tight_layout(pad=1.5)

```
